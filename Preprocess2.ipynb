```python
def basic_clean(data):
  data = data.replace("1", " ")
  data = data.replace("2", " ")
  data = data.replace("3", " ")
  data = data.replace("4", " ")
  data = data.replace("5", " ")
  data = data.replace("6", " ")
  data = data.replace("7", " ")
  data = data.replace("0", " ")
  data = data.replace("8", " ")
  data = data.replace("9", " ")
  data = data.replace(". .", ".")
  data = data.replace("hath", "has")
  data = data.replace(". .", ".")
  data = data.replace('D.M.', '')
  data = data.replace(". .", ".")
  data = data.replace("\n", " ")
  data = data.replace("\'s", " ")
  data = data.replace('" ,"', " ")
  data = data.replace(',"', " ")
  data = data.encode('ascii', 'ignore').decode()
  stopwrds = stopwords.words('english')
  stopwrds.remove('not')
  stopwrds.remove('no')
  stopwrds.append('.')
  stopwrds.append('shall')
  stopwrds.append('thus')
  data = ' '.join([map_old_new.get(i, i) for i in data.split()])
  # print(data)
  return data
```
