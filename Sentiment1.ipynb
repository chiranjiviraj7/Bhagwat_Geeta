```python
from transformers import BertTokenizer
from torch.utils.data import TensorDataset
```

```python
tokenizer = BertTokenizer.from_pretrained('bert-base-uncased', do_lower_case = True)
```

```python
encoder_adidevananda = tokenizer.batch_encode_plus(df['English Translation By Swami Adidevananda'].values,
                                           add_special_tokens = True,
                                            return_attention_masks = True,
                                           pad_to_max_length = True,
                                           max_length = 256,
                                           return_tensors = 'pt')
```

```python
input_ids_adidevananda = encoder_adidevananda['input_ids']
attention_masks_adidevananda = encoder_adidevananda["attention_mask"]
labels_adidevananda = torch.tensor(df.label.values)
```

```python
data_adidevananda = TensorDataset(input_ids_adidevananda,attention_masks_adidevananda,labels_adidevananda)
```

```python
from transformers import BertForSequenceClassification
model = BertForSequenceClassification.from_pretrained('bert-base-uncased',
                                     num_labels = 6,
                                     output_attentions = False,
                                     output_hidden_states =  False)
```

```python
from torch.utils.data import RandomSampler,SequentialSampler,DataLoader
```

```python
dataloader_adidevananda = DataLoader(
    data_adidevananda,
    sampler= RandomSampler(data_adidevananda),
    batch_size = 32
)
```

```python
from transformers import AdamW,get_linear_schedule_with_warmup
optimizer = AdamW(model.parameters(),lr = 1e-5,eps = 1e-8)
```

```python
import random
seed_val = 17
random.seed(seed_val)
np.random.seed(seed_val)
torch.manual_seed(seed_val)
torch.cuda.manual_seed_all(seed_val)
```

```python
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
model.to(device)
```

```python
model.load_state_dict(torch.load('/content/drive/My Drive/BERT_ft_epoch16.pth'))
```

```python
def evaluate(dataloader_val):
    model.eval()
    loss_val_total = 0
    predictions,true_vals = [],[]
    for batch in tqdm(dataloader_val):
        batch = tuple(b.to(device) for b in batch)
        inputs = {'input_ids':  batch[0],
                  'attention_mask':batch[1],
                  'labels': batch[2]
                 }
        with torch.no_grad():
            outputs = model(**inputs)
        loss = outputs[0]
        logits = outputs[1]
        loss_val_total +=loss.item()
        logits = logits.detach().cpu().numpy()
        label_ids = inputs['labels'].cpu().numpy()
        predictions.append(logits)
        true_vals.append(label_ids)
    loss_val_avg = loss_val_total/len(dataloader_val)
    predictions = np.concatenate(predictions,axis=0)
    true_vals = np.concatenate(true_vals,axis=0)
    return loss_val_avg,predictions,true_vals
```

```python
val_loss,predictions,true_vals = evaluate(dataloader_adidevananda)
```

```python
x = []
for i in predictions:
  x.append(i.argmax())
x
```
