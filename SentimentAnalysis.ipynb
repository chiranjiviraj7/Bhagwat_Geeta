```python
adidevananda_gandhi = []
purohit_gandhi = []
sivananda_gandhi = []
purohit_sivananda = []
adidevananda_sivananda = []
adidevananda_purohit = []
for i in range(1,19):
  adidevananda_gandhi.append(sum(adidevananda[adidevananda['Chapter'] == i]['label'] == gandhi[gandhi['Chapter'] == i]['label'])/len(adidevananda[adidevananda['Chapter'] == i]))
  adidevananda_purohit.append(sum(adidevananda[adidevananda['Chapter'] == i]['label'] == purohit[purohit['Chapter'] == i]['label'])/len(adidevananda[adidevananda['Chapter'] == i]))
  adidevananda_sivananda.append(sum(adidevananda[adidevananda['Chapter'] == i]['label'] == sivananda[sivananda['Chapter'] == i]['label'])/len(adidevananda[adidevananda['Chapter'] == i]))
  purohit_sivananda.append(sum(purohit[purohit['Chapter'] == i]['label'] == sivananda[sivananda['Chapter'] == i]['label'])/len(adidevananda[adidevananda['Chapter'] == i]))
  purohit_gandhi.append(sum(purohit[purohit['Chapter'] == i]['label'] == gandhi[gandhi['Chapter'] == i]['label'])/len(adidevananda[adidevananda['Chapter'] == i]))
  sivananda_gandhi.append(sum(sivananda[sivananda['Chapter'] == i]['label'] == gandhi[gandhi['Chapter'] == i]['label'])/len(adidevananda[adidevananda['Chapter'] == i]))
```

```python
plt.figure(figsize=(14, 6))
plt.plot([i for i in range (1,19)], adidevananda_purohit, marker='o', linestyle='dotted', label='adidevananda_purohit')
plt.plot([i for i in range (1,19)], adidevananda_sivananda, marker='o', linestyle='dotted', label='adidevananda_sivananda')
plt.plot([i for i in range (1,19)], adidevananda_gandhi, marker='o', linestyle='dotted', label='adidevananda_gandhi')
plt.plot([i for i in range (1,19)], purohit_sivananda, marker='o', linestyle='dotted', label='purohit_sivananda')
plt.plot([i for i in range (1,19)], purohit_gandhi, marker='o', linestyle='dotted', label='purohit_gandhi')
plt.plot([i for i in range (1,19)], sivananda_gandhi, marker='o', linestyle='dotted', label='sivananda_gandhi')
plt.title('Jaccard similarity between authors taken two at a time')
plt.xlabel('chapters')
plt.ylabel('similarity')
plt.legend()
plt.show()
```
